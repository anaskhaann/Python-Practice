Recursion is nothing but the function calling itself.
With the help of this we can create a loop without writing the loops explicitly.

- While approaching recursion we need to have a 2 things.

1. A Base case i.e A condition whose value we know
2. How can i break down the problem into smaller chunks.

#### Recursion works like stack, one by one we go down to base condition from top. And once the base condition is reached we then move up one by one. So if we have to implement recursion we have to use stack data structures.


```python
# Simple example of Iterative Vs Recusrive for multiplication

# This is the example of iterative
def multi(a,b):

    result=0

    for i in range(b):
        result = result +a
    return result

print(multi(4,5))
```

```python
# This is recursive

def multi_r(a,b):

    if b==1:
        return a
    else:
        return a + multi_r(a,b-1)
    
print(multi_r(4,3))
```

```python
# Factorial using Recursion

# 5! = 5 * 4! 

def fact(num):
    if num == 1 or num ==0:
        return 1
    else:
        return num * fact(num-1)

print(fact(5))
```

```python
# Palindrome with recursion

def palin(text):

    if len(text)<=1:
        print("Palindrome")
    else:
        if text[0] == text[-1]:
            palin(text[1:-1])
        else:
            print("Not a Palindrome")

palin("kell")
palin("abba")
palin("madam")
palin("racing car")
```

### The Rabit Problem:

If 2 newborn rabbits are put in a pen, how many rabbits will be in the pen after 1 year?
Assume that rabbits:

- Always produce one male and one female offspring
- Can reproduce once every month
- Can only reproduce once they are one month old
- They Never die


```python
# Fibonnacii series

def fib(num):
    if num ==0 or num ==1:
        return 1
    else:
        return (fib(num-1) + fib(num-2))
    
print(fib(12))
# print(fib(24))
# print(fib(36)) #this will take time
# print(fib(48)) #this will be menace to time complexity

# Thus this code is highly in efficient
```

#### Reason for In efficiency




To reduce this Time Complexity by recalculating the value, we can use extra space by storing the values of already calculated series, and then used those values whenever needed.

This concept is know as memoization, in which we do space time trade off, by utilizing space we are reducing our time.


```python
# Fibb with memoization
# First we need to maintain a memory(we will use dictionary)

d={0:1, 1:1}

def fibb_with_memo(num,dic):
    """
    This function take a number and a dictionary as input.
    It calculates the fibonacci value of a given number using memoization technique.
    It returns the fibonacci value of a number.
    """
    if num in dic:
        return dic[num]
    else:
        # Calculating new number and storing it in dictionary for not calculating further
        dic[num]=(fibb_with_memo(num-1,dic) + fibb_with_memo(num-2,dic))
        return dic[num]
    
print(fibb_with_memo(12,d))
print(fibb_with_memo(24,d))
print(fibb_with_memo(36,d)) 
print(fibb_with_memo(48,d))
```
