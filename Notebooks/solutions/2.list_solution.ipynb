###`Problem 3:` Update no of items available

Suppose you are given a list of candy and another list of same size representing no of items of respective candy.

i.e -

```
candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]
```

Write a program to show no. of items of each candy type.

`Output:`

```
Jelly Belly-10
Kit Kat-20
Double Bubble-34
Milky Way-74
Three Musketeers-32

```


````python
# Write code here

candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]

result = [f"{i}-{j}" for i,j in zip(candy_list,no_of_items)]

for i in result:
    print(i)
````

###`Problem 4:` Running Sum on list

Write a program to print a list after performing running sum on it.

i.e:

`Input:`

```
list1 = [1,2,3,4,5,6]
```

`Output:`

```
[1,3,6,10,15,21]
```

#### Solution Logic




````python
# Write code here

l= [1,2,3,4,5,6]
sum=0
result=[]

# with range
# for i in range(len(l)):
#     sum = sum+l[i]
#     result.append(sum)
    
# Without Range
for i in l:
    sum = sum+i
    result.append(sum)

print(result)

"""
Range--> We are looping on the index thats why we are accessing elements of that index.
Without Range --> we are directly accessing the elements thus l[i] should not be passed we can directly pass the element i.
"""
````
